```python
import pandas as pd
import folium
from folium import plugins

import matplotlib as mpl
import matplotlib.pyplot as plt
import json #json파일 읽기
```

```python
help(folium.Icon)
```

```python
state_geo = '/content/drive/My Drive/롯데글로벌로지스/state_geo.geojson' #json 파일 객체 생성
state_geo2=json.load(open(state_geo, encoding='utf-8'))
```

```python
m=folium.Map(location=[35.8, 127.6], zoom_start=8) #지도 생성
```

```python
folium.Choropleth(geo_data=state_geo2, #geo_data에 json 객체 대입
                  key_on='feature',
                  fill_color='darkpurple',
                  fill_opacity=0.5,
                  line_opacity=0.5
                  ).add_to(m)
#choropleth 맵은 색상 또는 패턴을 사용하여 특정 통계에 대한 데이터를 사전 정의 된 영역과 관련시키는 맵 유형
m
```
